```python
import geopandas as gpd

import datashader as ds
import plotly.express as px
import plotly.graph_objects as go

import json
```

```python
faults = gpd.read_file('hazfaults2014_proj.shp')
faults['geometry'] = faults['geometry'].to_crs(epsg=4326)
```

```python
faults["DISP_SLIP_"].unique()
```

```python
faults_thrust = faults[faults["DISP_SLIP_"]=='Thrust']
faults_normal = faults[faults["DISP_SLIP_"]=='Normal']
faults_strike = faults[faults["DISP_SLIP_"]=='Strike Slip']
faults_unassi = faults[faults["DISP_SLIP_"]=='Unassigned']
```

```python
faults_thrust.to_file('hazfaults2014_proj_thrust.geojson', driver='GeoJSON')
```

```python
faults_normal.to_file('hazfaults2014_proj_normal.geojson', driver='GeoJSON')
```

```python
faults_strike.to_file('hazfaults2014_proj_strikeslip.geojson', driver='GeoJSON')
```

```python
faults_unassi.to_file('hazfaults2014_proj_unassigned.geojson', driver='GeoJSON')
```

```python
with open('hazfaults2014_proj_thrust.geojson') as json_file:
    cafault_thrust_json = json.load(json_file)
```

```python
with open('hazfaults2014_proj_normal.geojson') as json_file:
    cafault_normal_json = json.load(json_file)
```

```python
with open('hazfaults2014_proj_strikeslip.geojson') as json_file:
    cafault_strikeslip_json = json.load(json_file)
```

```python
with open('hazfaults2014_proj_unassigned.geojson') as json_file:
    cafault_unassigned_json = json.load(json_file)
```

```python
sf = pd.DataFrame()
sf['latitude'] = [37.7739]
sf['longitude'] = [-122.4312]
```

```python
fig = px.scatter_mapbox(sf, lat='latitude', lon='longitude', zoom=4, opacity=1)
fig.update_layout(
                 mapbox_style="carto-darkmatter",
                 mapbox_layers = [
                {
                    "sourcetype": "geojson",
                    "sourcelayer": "fault",
                    "type": "line",
                    "color": "blue",
                    "opacity": 0.8,
                    "source": cafault_normal_json,
                },
                 {
                    "sourcetype": "geojson",
                    "sourcelayer": "fault",
                    "type": "line",
                    "color": "yellow",
                    "opacity": 0.8,
                    "source": cafault_thrust_json,
                },
                 {
                    "sourcetype": "geojson",
                    "sourcelayer": "fault",
                    "type": "line",
                    "color": "red",
                    "opacity": 0.8,
                    "source": cafault_strikeslip_json,
                },
                 {
                    "sourcetype": "geojson",
                    "sourcelayer": "fault",
                    "type": "line",
                    "color": "green",
                    "opacity": 0.8,
                    "source": cafault_unassigned_json,
                }]
)
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
